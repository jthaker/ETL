```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file = "database.csv"
filedf = pd.read_csv(file)
```

```python
#filter out only 2011 data to match weapon data csv
filedf.loc[filedf['Year'] == 2011]
```

```python
cleanfiledf = filedf.loc[:, ["State", "Crime Type", "Victim Sex", "Victim Age", "Perpetrator Sex", "Perpetrator Age", "Relationship", "Weapon"]]

#rename columns
cleanfiledf = cleanfiledf.rename(columns={"Crime Type": "crime_type",
                                                          "Victim Sex": "victim_sex",
                                                          "Victim Age": "victim_age",
                                                        "Perpetrator Sex":"perpetrator_sex",
                                                         "Perpetrator Age": "perpetrator_age"})

cleanfiledf.head()

```

```python
weapondata = "2011data.csv"
file2df = pd.read_csv(weapondata)
file2df.head()
#file2df.dropna(inplace = True)
#file2df.head()
#file2df.dropna(how = 'any').count()
#file2df.dtypes
```

```python
file2dfnew = file2df.loc[:, ["Circumstances", "Handguns", "Rifles", "Shotguns", "Blunt\nobjects\n(clubs,\nhammers,\netc.)", "Knives or\ncutting\ninstruments", 
                            "Personal\nweapons\n(hands,\nfists, feet,\netc.)", "Poison", "Pushed\nor\nthrown\nout\nwindow", 
                             "Explosives", "Fire", "Narcotics", "Drowning", "Strangulation", "Asphyxiation"]]

#rename columns
file2dfnew = file2dfnew.rename(columns={"Blunt\nobjects\n(clubs,\nhammers,\netc.)": "blunt_object",
                                                          "Knives or\ncutting\ninstruments": "knives",
                                                          "Personal\nweapons\n(hands,\nfists, feet,\netc.)": "martial_arts",
                                                            "Pushed\nor\nthrown\nout\nwindow": "defenestration"})

#deleted total rows
weapons_df = file2dfnew.drop([0,1, 27,28, 29, 30, 31])
weapons_df.head()
```

```python
relationshipdata = "relationship2011.csv"
relationshipdf = pd.read_csv(relationshipdata)
relationshipdf.head()
```

```python
relationshipdfnew = relationshipdf.loc[:, ["Circumstances", "Husband", "Wife", "Mother", "Father", "Son", "Daughter", "Brother", "Sister", "Other\nfamily", "Acquaintance","Friend", "Boyfriend", "Girlfriend", "Neighbor", "Employee", "Employer", "Stranger"]]

relationshipdfnew = relationshipdfnew.rename(columns={"Other\nfamily": "other_family"})



relationshipdfnew.head()
```
